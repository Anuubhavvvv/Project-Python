```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()
x = iris.data[:,:2] # independent features
y = iris.target     # dependent data - 3 categories
```

```python
print(x.shape)
print(y.shape)
```

```python
x[:5]
```

```python
y
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
from sklearn.svm import SVC, SVR
```

### Plotting SVM with different kernels

```python
kernel =['linear','rbf','poly']
for i in kernel:
    model = SVC(kernel=i) # default C=1.0
    model.fit(x,y)
    plot_decision_regions(x, y, clf=model)
    plt.xlabel('Sepal Length')
    plt.ylabel('Sepal Width')
    plt.title(f'SVM with kernel {i}')
    plt.show()
```

### Plot rbf kernel with different gamma

```python
g = [0.01,0.1,1,10,100]
for i in g:
    model = SVC(kernel='rbf',gamma=i) # default C=1.0
    model.fit(x,y)
    plot_decision_regions(x, y, clf=model)
    plt.xlabel('Sepal Length')
    plt.ylabel('Sepal Width')
    plt.title(f'SVM with rbf kernel  with gamma={i}')
    plt.show()
```

### Plot poly kernel with different C

```python
c = [0.01,0.1,1,10,100]
for i in c:
    model = SVC(kernel='poly',C=i) # default C=1.0
    model.fit(x,y)
    plot_decision_regions(x, y, clf=model)
    plt.xlabel('Sepal Length')
    plt.ylabel('Sepal Width')
    plt.title(f'SVM with poly kernel with C={i}')
    plt.show()
```

### Plot rbf kernel with different C

```python
c = [0.01,0.1,1,10,100]
for i in c:
    model = SVC(kernel='rbf',C=i) # default C=1.0
    model.fit(x,y)
    plot_decision_regions(x, y, clf=model)
    plt.xlabel('Sepal Length')
    plt.ylabel('Sepal Width')
    plt.title(f'SVM with rbf kernel with C={i}')
    plt.show()
```

```python
### Plot linear kernel with different C
```

```python
c = [0.01,0.1,1,5,10,30,60,100,300]
for i in c:
    model = SVC(kernel='linear',C=i) # default C=1.0
    model.fit(x,y)
    plot_decision_regions(x, y, clf=model)
    plt.xlabel('Sepal Length')
    plt.ylabel('Sepal Width')
    plt.title(f'SVM with linear kernel with C={i}')
    plt.show()
```

### Plot poly kernel with different degree

```python
deg = [1,2,3,4]
for i in deg:
    model = SVC(kernel='poly',degree=i) # default C=1.0
    model.fit(x,y)
    plot_decision_regions(x, y, clf=model)
    plt.xlabel('Sepal Length')
    plt.ylabel('Sepal Width')
    plt.title(f'SVM with poly kernel with degree={i}')
    plt.show()
```
